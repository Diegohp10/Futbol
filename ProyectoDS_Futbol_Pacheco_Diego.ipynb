# SIGLAS DE LAS COLUMNAS, EXPLICACION

**Significado de las siglas en los nombres de las columnas**  
   
Player: The name of the player.  
Nation: The player's nationality.  
Pos: The player's position (e.g., forward, midfielder, defender).  
Age: The player's age.  
MP (Minutes Played): Total minutes played by the player.  
Starts: Number of matches the player started.  
Min (Minutes): Total minutes played by the player (this might be the same as MP).  
90s (90s Played): The equivalent of 90-minute matches played by the player (e.g., 1.5 = 135 minutes).  
Gls (Goals): Total number of goals scored by the player.  
Ast (Assists): Total number of assists made by the player.  
G+A (Goals + Assists): Total number of goals and assists combined.  
G-PK (Goals - Penalty Kicks): Total number of goals scored excluding penalty kicks.  
PK (Penalty Kicks): Number of penalty goals scored by the player.  
PKatt (Penalty Kicks Attempted): Number of penalty kicks attempted by the player.  
CrdY (Yellow Cards): Number of yellow cards received by the player.  
CrdR (Red Cards): Number of red cards received by the player.  
xG (Expected Goals): The expected number of goals from the player's shots.  
npxG (Non-Penalty Expected Goals): Expected goals excluding penalties.  
xAG (Expected Assists): The expected number of assists from the player's passes.  
npxG+xAG (Non-Penalty xG + xAG): Total of non-penalty expected goals and expected assists.  
PrgC (Progressive Carries): Number of times the player carried the ball forward.  
PrgP (Progressive Passes): Number of passes made by the player that moved the ball forward.  
PrgR (Progressive Runs): Number of times the player made runs forward with the ball.  
Ast (Assists): (Repeated, already defined) Total number of assists made.  
G+A (Goals + Assists): (Repeated, already defined) Total number of goals and assists combined.  
G-PK (Goals - Penalty Kicks): (Repeated, already defined) Goals scored excluding penalty kicks.  
G+A-PK (Goals + Assists - Penalty Kicks): Total goals and assists minus penalty goals.  
xG (Expected Goals): (Repeated, already defined) Expected number of goals from the player's shots.  
xAG (Expected Assists): (Repeated, already defined) Expected number of assists from the player's passes.  
xG+xAG (Expected Goals + Expected Assists): Total expected goals and assists.  
npxG (Non-Penalty Expected Goals): (Repeated, already defined) Expected goals excluding penalties.  
npxG+xAG (Non-Penalty xG + Expected Assists): Total of non-penalty expected goals and expected assists.

** CALCULAR Y MOSTRAR GOLEADORES DE LA TEMPORADA Y ESTIMAR EN BASE A LA MUESTRA EL MEJOR CANDIDATO PARA CONTRATARLO A MI EQUIPO **

```python
df_Y = df_Fut.groupby(['Player', 'Squad'])['Gls'].mean()[df_Fut.groupby(['Player','Squad'])['Gls'].mean() > 20].reset_index()
plt.figure(figsize=(12,8))
sns.barplot(y='Gls',x='Player', hue='Squad', data=df_Y)
plt.title('Jugadores de mejor Promedio de Gol X Partido')
plt.yticks([0,5,10,15,20,25,30,35,40,45,50])
plt.xticks(rotation= 45)
plt.ylabel('Promedio de Gol')
plt.xlabel('Jugador')
plt.show()
```

** VIENDO LOS DATOS DE LOS MAYORES ASISTIDORES EN UNA MUESTRA GRAFICA DECIDIR QUE JUGADOR ES EL MAS IMPORTANTE Y ACONSEJAR CONTRATACION **

```python
df_Y2 = df_Fut.groupby(['Player'])['Ast'].sum()[df_Fut.groupby(['Player'])['Ast'].sum() > 11].reset_index()
df_Y2
```

```python
fig,ax= plt.subplots (figsize =(8,9))
plt.pie(df_Y2['Ast'], labels=df_Y2['Player'], autopct=lambda p: '{:.0f}'.format(p * sum(df_Y2['Ast']) / 100))
plt.title ('Jugadores con mas Asistencias')
plt.show()
```

** SABIENDO CUAL ES EL VALOR DE LOS MINUTOS JUGADOS POR LOS DATOS RECOGIDOS EN LA BASE SABREMOS QUE CANTIDAD DE MINUTOS ESTAN JUGANDO PARA PODER DETERMINAR SI HAY UNA SOBREEXIGENCIA Y AMENGUAR LAS CARGAS A FUTURO **

** SABER DE LA CANIDAD DE VECES QUE LOS JUGADORES CONVIERTEN LOS PENALES NOS SIRVE PARA TENER UNA MUESTRA DE COMO SON DE EFECTIVOS LOS JUGADORES PATEANDO PENALES**

```python
df_Y3 = df_Fut.groupby(['Player'])['PK'].sum()[df_Fut.groupby(['Player'])['PK'].sum() > 2].reset_index()
plt.figure(figsize=(12,8))
plt.hist(df_Y3['PK'], bins=10, edgecolor='black')
plt.title('Histograma de goles realizados por Penal')
plt.xlabel('PK')
plt.ylabel('Frecuencia')
plt.show()
```

**Obtener una muestra de la cantidad e goles convertidos segun la edad del jugador nos permite tener una idea de la tendencia de que rango etario es mas efectivo y tomar desicion de que tipo de jugador buscar para contratar segun la edad**
*EN EL GRAFICO SE PUEDE VISUALIZAR UN OUTLIER EN EL AÑO 2000 DONDE SOLO UN JUGADOR PARTICIPO EN MAS DE 50 PARTIDOS ALEJADO DE LA MEDIA*

```python
df_scatter = df_Fut.groupby(['Player', 'Born'])['MP'].sum().reset_index()
df_scatter = df_scatter[df_scatter['MP'] > 0]
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Born', y='MP', data=df_scatter)
plt.title('Cantidad de partidos jugados segun el año de nacimiento')
plt.ylabel('Partidos Jugados')
plt.xticks(rotation=45)
plt.show()
```

**PODEMOS SABER QUE JUGADORES TIENEN MEJOR PROYECCION DE GOLES SIN CONTAR LOS PENALES **

```python
df_Y4 = df_Fut.groupby(['Player'])['npxG'].sum()[df_Fut.groupby(['Player'])['npxG'].sum() > 15].reset_index()
plt.figure(figsize=(12,8))
sns.barplot(y='npxG',x='Player', data=df_Y4)
plt.title('Los mejores 15 jugadores con mejor espectativa de goles excuyendo penales')
plt.xticks(rotation= 45)
plt.ylabel('Exp Gol sin penales')
plt.xlabel('Jugador')
plt.show()
```
